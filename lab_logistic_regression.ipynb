```python
# prep: import modules and get pwd
import pymysql
from sqlalchemy import create_engine
import pandas as pd
import numpy as np
import getpass  # To get the password without showing the input
password = getpass.getpass()

pd.set_option('display.max_columns', None) # Used to show all columns
```

## Loading data

#### Data is loadaded after querying in MySQL.

```python
connection_string = 'mysql+pymysql://root:' + password + '@localhost/sakila'
engine = create_engine(connection_string)
query = '''SELECT * FROM sakila.film f
JOIN sakila.inventory i USING (film_id)
JOIN sakila.rental fa USING (inventory_id)
JOIN sakila.film_category fc USING (film_id)
JOIN sakila.category c USING (category_id)
JOIN sakila.payment p USING (rental_id);'''

sakila = pd.read_sql_query(query, engine)
sakila.head()
```

#### Checking data

```python
sakila.shape
sakila.dtypes
sakila.isna().sum()
sakila.drop_duplicates()
sakila.head(2)
```

#### Isolating the rental_date month in a separate column that we call 'may_target'

```python
sakila['may_target'] = pd.DatetimeIndex(sakila['rental_date']).month
sakila.head(2)
```

#### Create a new column for duration of customer rental which is return date - rental date (a timedelta64 dtype)

```python
sakila['duration_of_rental'] = (sakila['return_date'] - sakila['rental_date'])

sakila.head() # This returns 184 NaTs in the new column

sum(sakila['duration_of_rental'].isna())

# Replace the NaTs with the column median 

time_median = sakila['duration_of_rental'].median()
time_median
sakila['duration_of_rental'] = sakila['duration_of_rental'].replace({np.nan: time_median}).dt.days

```

#### Function: create a new column to convert our target into a boolean. If month is may return 1, else return 0.

```python
def boolean(x):
        
        if x == 5:
            return 1
        else:
            return 0

sakila['may_target'] = list(map(boolean, sakila['may_target']))
sakila.head(10)
```

#### Dropping columns that we will not be using and reordering the remaining ones.

```python
sakila = sakila.drop(['category_id', 'rental_id', 'inventory_id', 'description', 'release_year', 'language_id', 'original_language_id', 'last_update',
                     'last_update', 'rental_date', 'customer_id', 'return_date', 'staff_id', 'replacement_cost', 'payment_id', 'payment_date'], axis=1)
```

```python
sakila = sakila[['film_id', 'title', 'rental_duration', 'rental_rate', 'length',
       'rating', 'special_features', 'store_id', 'name', 'amount', 'duration_of_rental', 'may_target']]
```
